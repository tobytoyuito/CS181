```python
import numpy.random as npr
import numpy as np
import sys
import cPickle as pickle
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
record = pickle.load(open('record.p','rb'))
```

```python
group = [[5,5,3], [10, 10, 3], [10, 10, 5], [5, 5, 5]]
colors = ['r', 'b', 'g', 'yellow']
```

```python
for i in xrange(len(group)):
    plt.plot(record[i], label='Group %d'%(i+1), color = colors[i])
    plt.title('Score per Epoch')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()
```

```python
for i in xrange(len(group)):
    plt.plot(np.cumsum(record[i])/np.array(range(1,201)), label='Group %d'%(i+1), color = colors[i])
    plt.title('Average Score per Epoch')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()
```
