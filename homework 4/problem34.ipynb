```python
d=3
x = np.linspace(0,6,100)
samples = np.random.multivariate_normal(mean=[0]*d, cov = np.identity(d), size=10000)
samples = np.sqrt(np.sum(samples**2, axis=1))
y = scipy.stats.chi.pdf(x,d)
plt.hist(samples, normed=True, bins=50, label='Samples')
plt.plot(x, y, label='$\chi$ distribtuion with 3 df')
plt.title('True distribution v.s. Samples')
plt.legend()
plt.show()
```

```python
def KMean(X, k):
    N = len(X)
    J=[]
    choice=np.random.choice(range(N), size=k, replace=False)
    mu = X[choice]
    all_distance = np.zeros((N,k))
    for i in xrange(k):
        all_distance[:,i] = np.sum((X-mu[i])**2, axis=1)
    J.append(np.sum(np.min(all_distance, axis = 1)))
    
    for j in xrange(300):
        r = np.zeros((N,k))
        for i in xrange(N):
            r[i,np.argmin(all_distance[i])]=1

        for i in xrange(k):
            mu[i] = np.mean(X[r[:,i]==1], axis=0)
        
        for i in xrange(k):
            all_distance[:,i] = np.sum((X-mu[i])**2, axis=1)
            
        J.append(np.sum(np.min(all_distance, axis = 1)))
        if(J[j]==J[j-1]):
            break
        
        if not j%10:
            print j
    
    return mu, J, r
```

```python
def KMean_plus(X, k):
    N = len(X)
    choice=np.random.choice(range(N), size=1)
    mu = np.zeros((k,X.shape[1]))
    mu[0]=X[choice]
    J=[]

    for i in xrange(k-1):
        all_distance = np.zeros((N,len(choice)))
        for j in xrange(len(choice)):
            
            all_distance[:,j] = np.sum((X-mu[j])**2, axis=1)
        prob = np.min(all_distance, axis = 1)
        prob = prob/np.sum(prob)
        choice = np.random.choice(range(N), size=1, p=prob)
        mu[i+1] = X[choice]
    
    all_distance = np.zeros((N,k))
    for i in xrange(k):
        all_distance[:,i] = np.sum((X-mu[i])**2, axis=1)
    J.append(np.sum(np.min(all_distance, axis = 1)))
    
    for j in xrange(300):
        r = np.zeros((N,k))
        for i in xrange(N):
            r[i,np.argmin(all_distance[i])]=1

        for i in xrange(k):
            mu[i] = np.mean(X[r[:,i]==1], axis=0)
        
        for i in xrange(k):
            all_distance[:,i] = np.sum((X-mu[i])**2, axis=1)
            
        J.append(np.sum(np.min(all_distance, axis = 1)))
        if(J[j]==J[j-1]):
            break
        
        if not j%10:
            print j
    
    return mu, J, r
```
